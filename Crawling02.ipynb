```python
# 원래는 pip install requests 로 깔아주어야 함 
import requests # 웹 연결 라이브러리 
```

```python
from bs4 import BeautifulSoup as bs # 데이터를 분석하기 쉽게 정제 라이브러리 
```

```python
import pandas # 데이터 분석 라이브러리 
```

```python
url = "https://music.bugs.co.kr/chart"
html = requests.get(url)
html.text
```

```python
soup = bs(html.text, "html.parser")
```

```python
len(soup.select("tr")) #tr을 몽땅 선택하는 느낌~ 
```

```python
len(soup.select("table > tbody > tr"))
```

```python
songs = soup.select("table.byChart > tbody > tr")
```

```python
song = songs[0] #index 번호로 지정 ~ 

```

```python
# a tag는 link로 걸려 있기 때문에 대부분 a tag~로 접근하는 방식 
singer = song.select("p.artist > a")[0].text
singer
```

```python
title = song.select("p.title > a")[0].text
title 
```

```python
song_data = [] #빈 리스트~ 리스트로 담아야 다른 자료형으로 변환하기 쉬움!
rank = 1
songs = soup.select("table.byChart > tbody > tr")
for song in songs:
  title = song.select('p.title > a')[0].text 
  singer = song.select("p.artist > a")[0].text 

  song_data.append([rank, title, singer])
  rank += 1 # ramk = rank +1 



```

```python
import pandas as pd 
df = pd.DataFrame(song_data, columns = ["순위", "타이틀", "가수"])
df
```

```python
df.to_excel("bugs.xlsx", index=False) # 인덱스를 빼고 엑셀로 저장 
```
